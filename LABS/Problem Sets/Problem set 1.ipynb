```python
import networkx as nx
import matplotlib.pyplot as plt

import numpy as np
import random as rd
```

# Problema 2

```python
G = nx.Graph()
```

```python
fig = plt.figure()
fig.set_size_inches(10, 10)

nodes = list(np.linspace(0, 6, 6))
nodes = [int(node) for node in nodes]

G.add_nodes_from(nodes)

edges = [(0, 2), (1, 2), (1, 3), (2, 3), (2, 5), (3, 4), (4, 5), (4, 6)]
G.add_edges_from(edges)

pos = {0: (-1.8, -1), 1: (-1, -2.7), 2: (-2.1, -2.2), 3: (-2, -3.3), 4: (-3, -3), 5: (-3.2, -2), 6: (-4, -4)}

options = {
    "font_size": 36,
    "node_size": 3000,
    "node_color": "black",
    "edgecolors": "white",
    "font_color": "white",
    "linewidths": 5,
    "width": 5,
}

nx.draw_networkx(G, pos, **options)
plt.axis(False)
```
