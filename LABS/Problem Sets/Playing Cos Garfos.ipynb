```python
import networkx as nx
import matplotlib.pyplot as plt

import numpy as np
import random as rd
```

Creates graph

```python
graph = nx.Graph()
```

function I'll call for drawing

```python
def draw():
    nx.draw(graph, with_labels=True, font_weight='bold', font_color='white')
```

## Nodes

Adds nodes

```python
# Adds node n
n = 2
graph.add_node(2)

draw()
```

```python
graph.clear()

# Adds nodes on list
nodes = [1, 2, 3, 4]
graph.add_nodes_from(nodes)

draw()
```

## Edges

```python
graph.add_edge(1, 2)

draw()
```

```python
graph.add_edges_from([(2, 3), (3, 4), (4, 1)])

draw()
```

```python
G = nx.Graph()
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(1, 5)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge(4, 5)

# explicitly set positions
pos = {1: (0, 0), 2: (-1, 0.3), 3: (2, 0.17), 4: (4, 0.255), 5: (5, 0.03)}

options = {
    "font_size": 36,
    "node_size": 3000,
    "node_color": "white",
    "edgecolors": "black",
    "linewidths": 5,
    "width": 5,
}
nx.draw_networkx(G, pos, **options)

# Set margins for the axes so that nodes aren't clipped
ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()
```

Random Graph with 100 nodes

```python
nodes = list(np.linspace(1, 100, 100))

graph.add_nodes_from(nodes)

for run in range(1):
    n_edges = int(100 * rd.random() + 100)
    for i in range(n_edges):
        graph.add_edge(i, int(rd.random()*100))
```

```python
fig = plt.figure()
fig.set_size_inches(10, 10)

options = {
    "node_size": 500,
    "node_color": "turquoise",
    #"node_color": "white",
    "edgecolors": "black",
    "edge_color": "salmon",
    "linewidths": 1,
    "width": 1,
}

nx.draw(graph, **options)

fig.savefig('fofo_3.pdf')
```
